```python
import geopandas as gpd
import momepy as mp
```

```python
ROAD_HIERARCHY_MAP = {
    "primary": 4,
    "secondary": 3,
    "tertiary": 2,
    "quaternary": 1,
}
BIKE_HIERARCHY_MAP = {
    "primary": 2,
    "secondary": 1,
}
```

```python
gdf_edges = gpd.read_file("../data/processed/bikenet_edges.gpkg")
```

```python
G = mp.gdf_to_nx(gdf_edges, integer_labels=False, preserve_index=True)
```

```python
gdf_edges.groupby("built_in")["length"].sum() / gdf_edges["length"].sum()
```

```python
gdf_edges["highway"] = gdf_edges["highway"].map(ROAD_HIERARCHY_MAP)
gdf_edges["level"] = gdf_edges["level"].map(BIKE_HIERARCHY_MAP)
```

```python
gdf_edges.groupby("highway")["length"].sum() / gdf_edges["length"].sum()
```

```python
gdf_edges.groupby("level")["length"].sum() / gdf_edges["length"].sum()
```

```python
gdf_edges[["highway", "level"]].corr()
```

```python
gdf_edges.groupby(["highway", "level"])["length"].sum() / gdf_edges["length"].sum()
```
