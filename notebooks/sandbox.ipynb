```python
import geopandas as gpd
import momepy as mp
import networkx as nx
```

```python
gdf = gpd.read_file("../data/processed/bikenet_edges.gpkg")
```

```python
coins = mp.COINS(gdf)
```

```python
continuity_graph = mp.coins_to_nx(coins)
```

```python
nx.set_node_attributes(
    continuity_graph,
    dict(nx.betweenness_centrality(continuity_graph)),
    "stroke_betweenness",
)
```

```python
nx.set_node_attributes(
    continuity_graph,
    dict(nx.closeness_centrality(continuity_graph)),
    "stroke_closeness",
)
```

```python
strokes, edges = mp.nx_to_gdf(continuity_graph)
```

```python
strokes = strokes.explode("edge_indices")
```

```python
strokes = strokes.set_index("edge_indices")
```

```python
gdf["stroke_betweenness"] = strokes["stroke_betweenness"]
gdf["stroke_closeness"] = strokes["stroke_closeness"]
```
