```python
import geopandas as gpd
import pandas as pd
```

```python
segment_matches = gpd.read_file(
    "../data/processed/map_matching/segment_matches_15_17_30.gpkg"
)
osm_segments = gpd.read_file("../data/processed/map_matching/osm_segments_10.gpkg")
```

```python
osm_segments
```

```python
osm_linestrings = gpd.read_file("../data/processed/paris_car_edges_neatified.gpkg")
```

```python
osm_linestrings
```

```python
# Join the two datasets, keeping only OSM geometry
gdf_joined = pd.merge(
    segment_matches,
    osm_segments,
    left_on="matches_id",
    right_on="seg_id",
    how="right",
)
gdf_joined = gdf_joined[
    ["street", "level", "constructed", "osmid", "highway", "geometry_y"]
]
gdf_joined = gdf_joined.rename({"geometry_y": "geometry"}, axis=1)
gdf_joined.to_file("../data/processed/map_matching/pes_to_osm_raw.gpkg")
```

```python
# Join the two datasets, keeping only OSM geometry
gdf_joined = pd.merge(
    segment_matches,
    osm_segments,
    left_on="matches_id",
    right_on="seg_id",
    how="left",
)
gdf_joined = gdf_joined[
    ["street", "level", "constructed", "osmid", "highway", "geometry_y"]
]
gdf_joined = gdf_joined.rename({"geometry_y": "geometry"}, axis=1)
gdf_joined.to_file("../data/processed/map_matching/pes_to_osm_raw.gpkg")
```

```python
# Join the two datasets, keeping only OSM geometry
gdf_joined = pd.merge(
    segment_matches,
    osm_segments,
    left_on="matches_id",
    right_on="seg_id",
    how="right",
)
gdf_joined = gdf_joined[gdf_joined["constructed"].notna()]
gdf_joined = gdf_joined[
    ["street", "level", "constructed", "osmid", "highway", "geometry_y"]
]
gdf_joined = gdf_joined.rename({"geometry_y": "geometry"}, axis=1)
gdf_joined.to_file("../data/processed/map_matching/osm_only_raw.gpkg")
```
